```python
############## PRECEPTRON_AND ###################
x1 = [0,0,1,1]
x2 = [0,1,0,1]
target = [0,0,0,1]
```

```python
weights = [1,1,1]
```

```python
lr_rate = 1
```

```python
epochs = 10
```

```python
def threshold(x):
    print("x: ",x)
    if x>= 0:
        return 1
    else:
        return 0
```

```python
for i in range(epochs):
    error = []
    print("iteration Number: ",i)
    k = 0
    for i in range(4):
        print("input data: ", x1[i], x2[i])
        s = 0
        s += 1*weights[0] + x1[i]*weights[1] + x2[i]*weights[2]
        result = threshold(s)
        print("obtained Output: ", result)
        error.append(target[i] - result)
        print("Error: ", error[i])
        if error[i] !=0:
            k+=1
            weights[0] += error[i]*1*lr_rate
            weights[1] += error[i]*x1[i]*lr_rate
            weights[2] += error[i]*x2[i]*lr_rate
        print("After updation: ")
        print(weights)
#     print(k)
    if(k==0):
        print("Model trained perfectly")
        print(error)
        break
```

```python
############## PRECEPTRON_OR ###################
target = [0,1,1,1]
w0=w1=w2=1
lr=1
threshold = 0.0


for i in range(10):
    print("Iteration: ", i+1)
    k=0
    for j in range(len(x1)):
        print("Input: ", x1[j], x2[j])
        if( w0*1+w1*x1[j]+w2*x2[j])>=threshold:
            output = 1
        else:
            output=0
        error = target[j]-output
        print("Error: ",error)
        if(error!=0):
            k+=1
            w0 = w0+lr*error*1
            w1 = w1+lr*error*x1[j]
            w2 = w2+lr*error*x2[j]
        
        print("Updated Weights:", w0,w1,w2)
#     print(k)
    if k==0:
        print("Model has been trained")
        print("Final Weights: ", w0,w1,w2)
        print("Iteration: ",i)
        break
            
                
```
